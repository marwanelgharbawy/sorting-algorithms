# 1. Insertion Sort

## 1.1 Implementation

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i
        while j > 0 and arr[j-1] > key:
            arr[j] = arr[j-1]
            j -= 1
        arr[j] = key
```

 ## 1.2 Calculating Time Algorithm

```python
import time


def calc_time(arr):
    # store starting time
    begin = time.time()

    # program body starts

    insertion_sort(arr)
    # program body ends

    time.sleep(1)
    # store end time
    end = time.time()

    time_taken = end - begin

    return time_taken
```


## 1.3 Calculating Times

```python
import random

sizes = [10, 50, 100, 250, 1000, 2500]
times = []
for size in sizes:
    # generating arrays
    arr = [random.randint(0, 100) for _ in range(size)]
    # calculating time and adding it to times array
    times.append(calc_time(arr))
times
```


## 1.4 Graph

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o')
plt.title('Insertion Sort Time Complexity')
plt.xlabel('Array Size (n)')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()
```
