# 1. Bubble Sort

## 1.1 Implementation

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

 ## 1.2 Calculating Time Algorithm

```python
import time
def calc_time(arr):
    # store starting time
    begin = time.time()

    # program body starts

    bubble_sort(arr)
    # program body ends

    time.sleep(1)
    # store end time
    end = time.time()

    time_taken = end - begin

    return time_taken
```

## 1.3 Calculating Times

```python
import random

sizes = [10, 50, 100, 250, 1000, 2500]
times = []
for size in sizes:
    # generating arrays
    arr = [random.randint(0, 100) for _ in range(size)]
    # calculating time and adding it to times array
    times.append(calc_time(arr))
times
```

## 1.4 Graph

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o')
plt.title('Bubble Sort Time Complexity')
plt.xlabel('Array Size (n)')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()
```
